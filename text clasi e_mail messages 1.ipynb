```python
punc = string.punctuation
punc
```

```python
def remove_punc(text):
    non_punc_text = "".join([char for char in text if char not in string.punctuation])
    return non_punc_text
```

```python
df['non_punc_text'] = df['body_text'].apply(lambda x: remove_punc(x.lower()))
```

```python
df.head(10)
```

```python
from nltk import word_tokenize
```

```python
def tokenize(text):
    tokens = word_tokenize(text)
    return tokens
```

```python
df['token_text'] = df['non_punc_text'].apply(lambda x: tokenize(x))
```

```python
df.head(10)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words_list = nltk.corpus.stopwords.words("english")
stop_words_list[:10]
```

```python
def remove_stop_words(tokenize_list):
    text = [word for word in tokenize_list if word not in stop_words_list]
    return text
```

```python
df['remove_stopword_text'] = df['token_text'].apply(lambda x: remove_stop_words(x))
```

```python
df.head()
```

```python
porter = nltk.PorterStemmer()
```

```python
def stemming(tokenised_text):
    text = [porter.stem(word) for word in tokenised_text]
    return text
```

```python
df['stemming_text'] = df['remove_stopword_text'].apply(lambda x:stemming(x))
```

```python
from nltk.stem import WordNetLemmatizer
lemma_nltk = WordNetLemmatizer()
```

```python
def lemma(text):
    text = [lemma_nltk.lemmatize(word) for word in text]
    return text
```

```python
df['lemma_text'] = df['remove_stopword_text'].apply(lambda x:lemma(x))
```
