```python
import pandas as pd
import numpy as np
import tensorflow as tf
import nltk
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
stop_words_list = nltk.corpus.stopwords.words("english")
lemma_nltk = WordNetLemmatizer()
```

```python
data = pd.read_csv("J:/Data science/data/NLP/disaster/train.csv")
```

```python
df = data.sample(frac=1 , random_state=42)
df.head()
```

```python
df.shape
```

```python
df.target.value_counts()
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = word_tokenize(text)
    text = " ".join([lemma_nltk.lemmatize(word) for word in tokens if word not in stop_words_list])
    return text
```

```python
df['text'] = df['text'].apply(lambda x:clean_text(x))
```

```python
x = df['text']
y = df['target']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train ,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization
text_vectorizer = TextVectorization(max_tokens=max_tokens,
                                   output_sequence_length=max_sent_length)
```

```python
text_vectorizer.adapt(x_train)
```

```python
from tensorflow.keras import layers
tf.random.set_seed(42)
embdding = layers.Embedding(input_dim=max_tokens,
                            input_length=max_sent_length,
                            output_dim = 64,
                           name="Embedding_1")
embdding
```

```python
import random
random_sent = random.choice(x_train)
samp_embd = embdding(text_vectorizer([random_sent]))
samp_embd
```

```python
samp_embd[0]
#لكل كلمه embedding خاص بيها 64
```

```python
inputs = layers.Input(shape=(1,) , dtype="string")
x = text_vectorizer(inputs)
print(x.shape)
x = embdding(x)
print(x.shape)
x = layers.GlobalAveragePooling1D()(x) #removes 10 (every time step)
print(x.shape)
outputs = layers.Dense(1,activation="sigmoid")(x)
print(outputs.shape)
model_1 = tf.keras.Model(inputs,outputs,name="model_1_dense")
```

```python
model_1.compile(loss=tf.keras.losses.binary_crossentropy,
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["accuracy"])
```

```python
model_1.summary()
```

```python
model_1_history = model_1.fit(x_train,y_train,epochs=10)
model_1_history
```

```python
model_1.evaluate(x_test,y_test)
```
